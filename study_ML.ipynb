### Понятние Производной


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
    return x**4 + 5*x**3 - 10*x

x = np.linspace(-5, 2, 100)
y = list(map(F, x))

plt.figure(figsize=(10, 7))
plt.plot(x, y)
plt.ylabel("Y")
plt.xlabel("X")
plt.scatter([-3.5518, -0.9439, 0.7457], [F(-3.5518), F(-0.9439), F(0.7457)], lw=5)
plt.show()
```

Для того чтобы измерять скорость возрастания или убывания функции, мы можем отметить на графике для какой-то из наших точек локального максимума или минимума, соответствующее приращение аргумента (дельта Х), и отметить приращение функции (дельта Y).

Отношение (дельта у) / (дельта х) показывает как быстро возрастает или убывает наша функция на участках (х, х+дельта).
Действительно это отношение показывает, на сколько в среднем меняется значение функции при изменении значения аргумента на 1.

### Чем больше это отношение по модулю, тем выше скорость убывания/возрастания функции

Выбирая фиксированное значени дельта х, мы рискуем тем, что в этот промежуток попадет наша точка экстремума, или же функция может возрастать с разной скоростью внутри нашего промежутка.

Чтобы эту проблему преодолеть, можно взять какое-то значение дельта х  и устремить его в пределе к нулю.
Другими словами будем уменьшать дельта х, и построим последовательность числе предел которой 0.

Теперь возьмем функцию F и некоторую точку этой функции (х, у). Рассмотрим приращение х+дельта и соответствующее приращение функции у + дельта = F(x+дельта).

Будем постепенно уменьшать приращение дельта х, подставляя вместо дельта х члены заданой последовательности и получим таким образом последовательность дельта у


дельта у1 = F(x + 1) - F(x)
дельат y2 = F(x + 1/2) - F(x) итд

Запишем отношение

дельта у1/ дельта х1 = F(x+1) - F(x)/ дельта х1  ИТД

Производная в точке определяется через предел
F'(x) = lim(delta x -> 0) F(x + delta x) - F(x)/ delta x

Когда такой предел существует, функция будет дифференцируема в точка (х, у)

Когда такого предела не существует, значит производной нет и функция не диффернцируема.


Если функция имеет такой предел для всех точек, такую функцию называем диффернцируемой всюду!!!

Также есть понятие чатсной производной для случая функции с n переменными.

Частная производная это производная когда диффернцирование происходит по одному выбраному аргументу, считая остальные переменные константами.


Правила взятия производной от произведния, деления, сложной функции итд такие же как для и функции с одной переменной.

(При желании можно изучить теорему шварца о частных производных)

### Градиент функции многих переменных - это вектор частных проивзодных функции по всем переменным

grad F = (F'_x1(x1, x2, ..., xn)), F'_x2(x1, x2, ..., xn), ..., F_xn(x1, x2, x3,..., xn)

### Градиент функции показывает, как ведет себя функция по каждой переменной.
Характер изменения функции по каждой переменной.

### Градиентная оптимизация


Постановка задачи:

Задача поиска минимума и максимума функции или задача оптимизации функции.

Эта задача решается нейронными сетями при обучении. Да, по сути, обучение нейронной сети - это оптимизация одной большой и очень сложной функции с тысячами или даже милионнами аргументов.

Чуть подробнее:

При обучении нейронной сети мы хотим измерять, насколько хорошо сеть справляется с поставленной задачей. Для этого вводится ФУНКЦИЯ ПОТЕРЬ - такая функция, которая в качестве аргументов принимает правильные ответы и ответы нейронной сети, и выдает некоторое число - ошибку.

Понятно, что мы хотим, чтобы ошибка сети была минимальной.

То есть, ставится задача минимизации функции потерь!!!

Мы хотим подобрать такие аргументы(параметры) нейронной сети, чтобы получить минимум функции потерь.


Существование экстремума:

1) Необходмое условие: Если функция F имеет экстремум в точке х, то Либо F не имеет производной в точке х, либо производная в точке х = 0 (F'(x) = 0)
2) Достаточное условие: Если функция F непрерывана и дважды диффернцируема в точке х_0. Тогда при условии 
F'(x_0) = 0 & F''(x_0) < 0 верно, что х_0 является точкой локального максимума.
А если F'(x_0) = 0 & F''(x_0) > 0 - является точкой локального минимума.

Иными словами, чтобы точка х была точкой экстремума важно выполнение двух условий.

### Алгоритм нахождения точек локальных минимумов функции F(x):

1) Найти корни уравнения F'(x) = 0 (Найти те точки х, в которых производная = 0)
2) Для всех корней уравнения, вычислить вторую производную F''(x).

Если она больше нуля, то это точка минимума. Если меньше нуля, то точка максимума

Алгоритм описанный выше, является аналитическим и дает сбой когда функция становится на столько сложной, что мы не можем ее вычислить аналитически.

Тогда нужна какая-то замена, алгоритм который не будет зависеть от сложности функции

### Алгоритм градиентного спуска: ИДЕЯ

Допустим мы выбрали какую-то точку на графике, взяли производную и видим что в этой точке функция возрастает.
Очевидно, что какая-то точка локального минимума функции находится левее точки х, которую мы взяли.
'левее' - означает, что минимум функции достигается при значении аргумента < х.

Это наблюдение значит, что если мы уменьшим  х на некотрую величину дельта(х), то мы можем стать ближе к точке минимума.

### Алгоритм поиска точки минимума
1) Берем случайную точку х функции F
2) Вычисляем производную F'(x)
3) Если F'(x) > 0, уменьшаем х, если F'(x) < 0, то увеличиваем х.
4) повторяем пункты 2, 3

Осталось два вопроса:
1) каким должно быть дельта(х) - то есть на сколько именно уменьшать
2) Когда остановить алгоритм ? 

Можем заметить, что когда мы приближаемся к точке минимума, тем плавнее становится график.

Формально, это означает уменьшение скорости возрастания/убывания функции, а значит тем меньше модуль значения производной.

Зафиксируем маленькое число Эпсилон(Е) и каждый шаг алгоритма будем двигать х по направлению к минимуму на шаг 
Е * |F'(x)|.
Тогда пока наша точка будет далеко от минимума, модуль производной будет довольно большим и мы будем двигаться большими шагами, а по мере приближения к точке минимума, модуль производной  будет уменьшаться и наш шаг тоже будет уменьшаться.

Так мы менее вероянто, перескочим через точку минимума и подойдем к ней ближе в итоге.

Ответ на второй вопрос -- когда остановить алгоритм поиска минимума -- может быть разный.
Действуя, таким образом, ровно в точку минимума мы почти никогда не придем.
Придется остановиться в какой-то точке возле минимума. Но для применения этого алгоритма нахождение точки около минимума вполне хватает. Чаще всего алгоритм останавливается после прохождения определенного количества шагов и/или достижения определенной близости к точке минимума. Близость, опять же, можно измерять значением производной в точке. Чем ближе, тем модуль производной меньше!!!

### Алгоритм градиентного спуска


1) Берем некоторую точку х функции F, фиксируем Е (for example E=0.001) Е- эпсилон (скорость обучения или learning item(rate))
2) Вычисляем производную F'(x)
3) Изменяем х: x = x - E * F'(x)
4) Повторяем 3, пока не пройдет определенное количество шагов и/или мы не станем достаточно близки к точке минимума 

### Почему называется ГРАДИЕНТНОГО

Потому что мы спускаемся к точке минимума, вычисляя производную (градиент) функции на каждом шаге.

### Оптимизация функции многих переменных
Идея нахождения точно такая же как и для одной переменной.

1) Берем некоторую точку х = x(x1, x2, ..., xn), функции F, фиксируем Е (Например Е = 0.001)
2) Вычисляем частные производные  F'(x) по всем n аргументам функции в точка x = (x1, x2, ..., xn), получаем градиент grad F = (F'_x1(x1, x2, ..., xn), F'_x2(x1, x2, ..., xn), ..., F'_xn(x1, x2, ..., xn))
3) Изменяем
    x1: x1 = x1 - E * F'_x1(x1, x2, ..., xn)
    x2: x2 = x2 - E * F'_x2(x1, x2, ..., xn)
    ...
    xn: = xn - E * F'_xn(x1, x2, ..., xn)

    или в векторной записи x = x - E * grad(F(x))
4) Повторяем 3, пока не пройдет определенное количество шагов и/или мы не станем достаточно близко к точке минимума по каждой из коориднат

Здесь мы пользуемся утверждением о том, что антиградиент задает направление наискорейшего убывания функции
